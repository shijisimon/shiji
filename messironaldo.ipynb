```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('MessiRonaldo.csv')
df.head()
```

```python
print(df.columns)
df.shape
```

```python
df.info()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#add the columns which show Mins per activities,Goals,Assists and Appearances.
df['Mins/Liga_Goals']=df['Liga_Mins']/df['Liga_Goals']
df['Mins/Liga_Asts']=df['Liga_Mins']/df['Liga_Asts']
df['Mins/Liga_Aps']=df['Liga_Mins']/df['Liga_Aps']
df['Mins/CL_Goals']=df['CL_Mins']/df['CL_Goals']
df['Mins/CL_Asts']=df['CL_Mins']/df['CL_Asts']
df['Mins/CL_Aps']=df['CL_Mins']/df['CL_Aps']
```

```python
df['lg_rank']=[1,3,2,1,1,2,1,2,2,3,1,2,2,1,1,2,1,2]
df['cl_rank']=[5,1,5,1,5,1,1,4,5,1,4,4,4,4,1,4,4,5]
```

```python
df.head()
```

```python
df=df.iloc[::-1]
df
```

```python
#compare Messi and Ronaldo in Total Goals, Assitants and Appearances
plt.figure(figsize=(15,15))
plt.subplot(421)
df.groupby('Player')['Liga_Goals'].sum().plot.bar()
plt.title('Liga_Goals')
plt.subplot(422)
df.groupby('Player')['Liga_Asts'].sum().plot.bar()
plt.title('Liga_Asts')
plt.subplot(423)
df.groupby('Player')['Liga_Aps'].sum().plot.bar()
plt.title('Liga_Aps')
plt.subplot(424)
df.groupby('Player')['Liga_Mins'].sum().plot.bar()
plt.title('Liga_Mins')
plt.subplot(425)
df.groupby('Player')['CL_Goals'].sum().plot.bar()
plt.title('CL_Goals')
plt.subplot(426)
df.groupby('Player')['CL_Asts'].sum().plot.bar()
plt.title('CL_Asts')
plt.subplot(427)
df.groupby('Player')['CL_Aps'].sum().plot.bar()
plt.title('CL_Aps')
plt.subplot(428)
df.groupby('Player')['CL_Mins'].sum().plot.bar()
plt.title('CL_Mins')
plt.subplots_adjust(hspace=1)
plt.subplots_adjust(wspace=1)
```

```python
Messi is more active in Liga than Ronalldo,and Ronaldo is more active in CL than Messi.
```
