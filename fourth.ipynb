# plot Pie chart using Seaborn

```python
import seaborn as sns
sns.set_theme(palette="dark",font="arial",font_scale=2)
data=[211000,183300,224700,222700,209600,201400,295500,361400,234000,267000,412800,310000]
months=["January","February","March","April","May","June","July","August","September","October","November","December"]
plt.pie(data,labels=months)
plt.show()
```
