# Reading a file

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('company_sales_data.csv')
print(df)
```

# Displot

```python
sns.displot(df, x="total_profit")
```

# Scatter Plot

```python
ax = df.plot.scatter(x="month_number", y="total_profit",figsize=(9,6),color='Red',title = 'Company profit per month')
```

# violin

```python
df1=sns.load_dataset('titanic')
df1
sns.violinplot(data=df1, x="deck", y="age", hue="alive")
```

# Histogram

```python
sns.histplot(data=df1, x="age", binwidth=9)
```

# Box and whisker plot

```python
sns.boxplot(data=df1, x='age', y='class',whis=[0.05, 0.95])
plt.show()
```
